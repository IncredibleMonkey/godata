# 미리 x_data, y_data를 입력받아서 코드작성 시험

```python
import tensorflow as tf
```

```python
# 예측 데이터, 실제 데이터 넣기 
x_data = [1, 2, 3]
y_data = [1, 2, 3]
```

```python
# hypothesis에 들어갈 변수 설정하기
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
# hypothesis로 그래프 만들기
hypothesis = x_data * W + b
```

```python
# cost function 설정하기
cost = tf.reduce_mean(tf.square(hypothesis - y_data))
```

```python
# 경사하강법 설정하기
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
# 최적의 W값을 찾기 : 
# (1) 총 2001번 실행 
# (2) 200번마다 결과화면에 <step, cost, W, b> 출력하기
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    sess.run(train)
    
    if (step % 200) == 0 :
        print(step, sess.run(cost), sess.run(W), sess.run(b))    
```

```python
# 그래프로 w값에 따른 cost값의 변화 나타내기
import matplotlib as plt
import seaborn as sns

W_val = []
cost_val = []

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(201):
    sess.run(train)
    
    W_val.append(sess.run(W))
    cost_val.append(sess.run(cost))

sns.scatterplot(x= W_val, y= cost_val)
```

# 플레이스홀더로 코드작성 시험

```python
X = tf.placeholder(dtype="float32")
Y = tf.placeholder(dtype="float32")

W = tf.Variable(tf.random_normal([1]), name = "weight")
b = tf.Variable(tf.random_normal([1]), name = "bias")
```

```python
hypothesis = W * X + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.01)
```

```python
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, W_val, b_val,_ = sess.run([cost, W, b, train], 
                                       feed_dict={X:[1,2,3,4,5], Y:[1,2,3,4,5]})
    
    print(cost_val, W_val, b_val)
```
